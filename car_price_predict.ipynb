```python
feature=['doornumber','carbody','enginetype','cylindernumber']
```

```python
df=df.drop(feature,axis=1)
```

```python
x=['doornumber', 'carbody', 'enginetype', 'cylindernumber']
categorical=[feat for feat in categorical if feat not in x ]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data=scaler.fit_transform(df[numerical])
```

```python
df1=pd.DataFrame(data,columns=numerical)
```

```python
df2=pd.concat([df1,df[categorical]],axis=1,join='inner')
```

```python
target=df2['price']
df2=df2.drop('price',axis=1)
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_mod=SelectFromModel(Lasso(alpha=0.002,random_state=0))
```

```python
feature_sel_mod.fit(df2,target)
```

```python
selectfeatures=feature_sel_mod.get_feature_names_out()
```

```python
selectfeatures
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df2[selectfeatures],target,random_state=0)
```

```python
xtrain.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain[selectfeatures],ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import r2_score
r2_score(model.predict(xtest),ytest)
```
