```python
def strip(x):
    return x.strip()



df['date']=df[' Date'].apply(strip)
df['date']=pd.to_datetime(df['date'],format="%d-%m-%Y")

```

```python
y=df[df['Region']=="Andhra Pradesh"]
y.set_index('date',inplace=True)
```

```python
z=y.resample("Y").agg({' Estimated Labour Participation Rate (%)': 'mean',' Estimated Unemployment Rate (%)':'mean'})
```
